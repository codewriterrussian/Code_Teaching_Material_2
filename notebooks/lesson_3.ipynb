1. 使用一行程式建立變數 name 和 age，分別存入你的名字與年齡，並輸出。
2. 使用一行程式建立 x = 8, y = 12，輸出 x + y。
3. 讓使用者輸入兩個數字，例如 7 9，分別存入 a, b，並輸出 a 和 b。
4. 承上題，將 a 和 b 轉成整數後，輸出 a + b。
5. 挑戰題：使用者輸入「飲料名稱 價格」，例如 milk_tea 60，請輸出「milk_tea 的價格是 60 元」。

```python
name, age = input("請輸入你的名字和年齡（用逗號分隔）：").split(",")
print("名字:", name, "年齡:", age)
```

```python
x, y = 8, 12
print(x+y)
```

```python
x, y = input("請輸入兩個數字（用逗號分隔）：").split(",")
x, y  = int(x), int(y)  # 把 x 和 y 從字串轉成整數
print(x+y)
```

```python
item, price = input("請輸入商品名稱和價格（用逗號分隔）：").split(",")

print(item, "的價格是", price, "元")
```

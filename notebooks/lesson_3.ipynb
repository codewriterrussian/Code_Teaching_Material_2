```python
print("2" * 2)      # 想說得到 "4"，但實際上得到 "22" (字串的重複)
print(2 * 2)        # 得到 4 (數字的乘法)
print(int("2") * 2) # 得到 4 (先把 "2" 轉成數字，再乘以 2)
```

```python
price = 60
count = '2'
total = price * float(count) 
# int 乘上 float -> int 自動轉成 float -> float 乘上 float -> float
print(total)
```

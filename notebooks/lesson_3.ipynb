## 1. `print()`：把資料顯示出來

最基本的輸出方式是使用 `print()`。

`print()` 可以一次印出多個資料，Python 會自動在資料中間加上一個空格。


````python
print("Hello", "Python")
print("Taipei", "Taiwan")
````

### 用 `sep` 改變中間的分隔符號

`sep` 是 separator 的意思，也就是「分隔符號」。

預設情況：

```python
print("A", "B", "C")
```

輸出會是：

```text
A B C
```

如果指定 `sep="/"`，中間就會改成 `/`。


````python
print("Taipei", "Taiwan", sep=", ")
print("A", "B", "C", sep="/")
print("2026", "06", "12", sep="-")
````

## 2. `end`：控制輸出結尾

`print()` 預設印完後會換行。

也就是：

```python
print("Hi")
print("Python")
```

輸出會分成兩行。

如果不想換行，可以設定 `end`。


````python
print("Hi")
print("Python")

print("Hi", end=" ")
print("Python")

print("A", end="")
print("B", end="")
print("C")
````

### 小整理：`sep` 和 `end` 的差別

| 參數 | 控制的位置 | 預設值 | 例子 |
|---|---|---|---|
| `sep` | 多個資料中間 | 空格 | `print("A", "B", sep="-")` |
| `end` | 整個輸出的最後 | 換行 | `print("A", end="")` |

記法：

```text
print(資料1, 資料2, sep="中間放什麼", end="最後放什麼")
```


## 3. `input()`：讓使用者輸入資料

`input()` 會等待使用者輸入文字。

括號裡可以放提示文字，讓使用者知道要輸入什麼。


````python
name = input("What is your name? ")
print("Hello", name)
````

### 輸入兩個資料

我們可以連續使用兩次 `input()`。


````python
name1 = input("First name: ")
name2 = input("Second name: ")

print(name1, name2, sep=" ")
````

### 範例：商品和價格

注意：這裡的 `price` 目前還是字串，不是數字。


````python
item = input("What is your favorite item? ")
price = input("How much does it cost? ")

print(item, price, sep=": ")
print(type(price))
````

## 4. `.split()`：把一串輸入切開

`.split()` 可以把一段文字切成多個部分。

例如：

```python
"John,Merry,Hank".split(",")
```

會變成：

```python
["John", "Merry", "Hank"]
```

也就是一個串列 list。


````python
names = "John,Merry,Hank,Petter,Harry".split(",")
print(names)
print(type(names))
````

### 從使用者輸入拆資料

請輸入類似：

```text
John,Merry,Hank
```


````python
names = input("請輸入多個名字，用逗號分隔：").split(",")
print(names)
````

### 用空格拆資料

如果 `.split()` 裡面不寫東西，預設會用空白來切。

請輸入類似：

```text
7 9
```


````python
a, b = input("請輸入兩個數字，用空格分隔：").split()
print(a)
print(b)
print(type(a), type(b))
````

## 5. 型態：文字和數字是不一樣的

Python 中常見的型態：

| 型態 | 名稱 | 例子 |
|---|---|---|
| `int` | 整數 | `10` |
| `float` | 小數 | `10.5` |
| `str` | 字串 | `"10"` |
| `bool` | 布林值 | `True`, `False` |
| `list` | 串列 | `[10, 20, 30]` |

可以用 `type()` 檢查資料型態。


````python
print(type(10))
print(type(10.0))
print(type("10"))
print(type(True))
print(type([10]))
````

### 字串的 `*` 和數字的 `*` 不一樣

這是初學者很常遇到的問題。

```python
"2" * 2
```

意思是把字串重複兩次，所以得到 `"22"`。

```python
2 * 2
```

才是數字乘法，得到 `4`。


````python
print("2" * 2)       # 字串重複
print(2 * 2)         # 數字乘法
print(int("2") * 2)  # 先轉成整數，再乘法
````

## 6. 型態轉換：`int()` 和 `float()`

如果想把輸入的文字拿來做數學運算，就要先轉成數字。

| 函式 | 作用 | 例子 |
|---|---|---|
| `int()` | 轉成整數 | `int("60")` → `60` |
| `float()` | 轉成小數 | `float("60.5")` → `60.5` |
| `str()` | 轉成字串 | `str(60)` → `"60"` |


````python
price = 60
count = "2"

total = price * int(count)
print(total)
print(type(total))
````

### 小數版本

如果數量可能是 `2.5`，可以用 `float()`。


````python
price = 60
count = "2.5"

total = price * float(count)
print(total)
print(type(total))
````

## 7. 多變數指定

Python 可以一行指定多個變數。


````python
a, b = 10, 20
print(a)
print(b)
````

### 搭配 `.split()` 使用

請輸入類似：

```text
milk_tea 60
```


````python
item, price = input("請輸入商品名稱和價格，用空格分隔：").split()

print(item, "的價格是", price, "元")
````

### 把拆出來的資料轉成數字

請輸入類似：

```text
7 9
```


````python
x, y = input("請輸入兩個數字，用空格分隔：").split()

print(type(x), type(y))

x = int(x)
y = int(y)

print(x + y)
print(type(x), type(y))
````

## 8. 串列 list：放很多資料的盒子

串列可以放多個資料，而且可以放不同型態。


````python
intro = ["lbj", 2, True]
print(intro)
print(type(intro))
````

### 串列也可以被拆成變數

右邊的資料數量要和左邊的變數數量一樣。


````python
a, b = [20, 30]
print(a)
print(b)
````

## 9. 變數更新

變數可以用自己原本的值更新。

```python
x = x + 10
```

意思是：

1. 先看右邊：`x + 10`
2. 算出新的值
3. 再放回左邊的 `x`


````python
x = 10
print("原本的 x:", x)

x = x + 10
print("更新後的 x:", x)
````

## 10. 常見錯誤提醒

### 錯誤 1：忘記轉型

```python
x, y = input().split()
print(x + y)
```

如果輸入：

```text
2 3
```

輸出會是：

```text
23
```

因為 `x` 和 `y` 都是字串。

---

### 正確做法

```python
x = int(x)
y = int(y)
print(x + y)
```


````python
x, y = input("請輸入兩個數字：").split()

print("還沒轉型：", x + y)

x = int(x)
y = int(y)

print("轉成數字後：", x + y)
````

## 11. 課堂練習

### 練習 1：自我介紹

請讓使用者輸入姓名、年齡、地區，並輸出一句完整句子。

輸入範例：

```text
Amy 13 Taipei
```

輸出範例：

```text
大家好，我是 Amy，今年 13 歲，來自 Taipei。
```


````python
# 練習 1：請學生自己完成
name, age, city = input("請輸入姓名 年齡 地區：").split()
print("大家好，我是", name, "今年", age, "歲，來自", city + "。")
````

### 練習 2：兩數相加

輸入兩個整數，輸出它們的總和。

輸入範例：

```text
8 12
```

輸出範例：

```text
20
```


````python
# 練習 2
x, y = input("請輸入兩個整數：").split()
x = int(x)
y = int(y)
print(x + y)
````

### 練習 3：商品總價

讓使用者輸入商品名稱、單價、數量，輸出總價。

輸入範例：

```text
tea 30 4
```

輸出範例：

```text
tea 總共 120 元
```


````python
# 練習 3
item, price, count = input("請輸入商品名稱 單價 數量：").split()
price = int(price)
count = int(count)

total = price * count
print(item, "總共", total, "元")
````

輸出範例：

```text

```


````python
# 練習 4
text = input("請輸入一段文字：")
print(text * 2)
````

## 12. 課後挑戰題

### 挑戰題 1：格式化日期

輸入年、月、日，輸出成 `年/月/日` 格式。

輸入範例：

```text
2026 6 12
```

輸出範例：

```text
2026/6/12
```


````python
# 挑戰題 1
year, month, day = input("請輸入 年 月 日：").split()
print(year, month, day, sep="/")
````

### 挑戰題 2：平均分數

輸入三個分數，輸出平均。

輸入範例：

```text
80 90 100
```

輸出範例：

```text
90.0
```


````python
# 挑戰題 2
s1, s2, s3 = input("請輸入三個分數：").split()
s1 = float(s1)
s2 = float(s2)
s3 = float(s3)

avg = (s1 + s2 + s3) / 3
print(avg)
````

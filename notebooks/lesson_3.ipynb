## 2. `input()` 與 `print(input())` 的差別

`input()`：讀取使用者輸入，並把輸入結果當成字串。  
`print(input())`：先讀取使用者輸入，再把輸入結果印出來。


```python
# 讀取輸入，並存進變數
name = input("請輸入你的名字：")
print("你好，", name)
```

```python
# 直接把輸入印出來
print(input("請輸入一句話："))
```

### 重要觀念：`input()` 得到的是字串 `str`

即使輸入的是數字，Python 也會先把它當成文字。  
如果要做數學計算，需要用 `int()` 或 `float()` 轉型。


```python
age = input("請輸入年齡：")

print(age)
print(type(age))
```

```python
age = int(input("請輸入年齡："))

print(age + 1)
print(type(age))
```

## 3. 布林值 Boolean：`True` / `False`

布林值常用在判斷：  
- `True`：條件成立  
- `False`：條件不成立  

注意：`True` 和 `"True"` 不一樣。  
`True` 是布林值，`"True"` 是字串。


```python
print(type(True))
print(type("True"))
```

### `bool()` 的常見結果

在 Python 裡，空字串、空串列、數字 `0` 通常會被判定為 `False`。  
非空字串即使內容是 `"False"`，也會被判定為 `True`。


```python
print(bool("False"))  # 非空字串，所以是 True
print(bool(""))       # 空字串，所以是 False
print(bool([]))       # 空串列，所以是 False
print(bool(0))        # 數字 0，所以是 False
print(bool(1))        # 非 0 數字，所以是 True
```

### 從一行輸入中切出多個資料

`split()` 可以把一段文字依照指定符號切開。  
例如：`"2026/05/18".split("/")` 會得到 `["2026", "05", "18"]`。

## 5. 序列 Sequence：程式由上到下執行

Python 會依照程式碼的順序執行。  
後面執行的程式可能會改變前面的變數結果。


```python
x = 10
print("第一次的 x：", x)

x = 20
print("第二次的 x：", x)
```

### 生活化理解：便條紙被改寫

變數像一張便條紙。  
一開始寫 `10`，後來改成 `20`，最後電腦看到的就是 `20`。


```python
wallet_amount = 1000
buy_drink = 150

wallet_amount = wallet_amount - buy_drink

print("買飲料後剩下：", wallet_amount)
```

### 練習：銀行帳戶更新

一開始銀行帳戶有 5000 元。  
接著：
1. 存入 500 元
2. 繳學費 1000 元
3. 收到薪水 2000 元

請算出最後帳戶剩下多少錢。


```python
bank = 5000
save = 500
tuition = 1000
salary = 2000

bank = bank + save - tuition + salary

print("最後帳戶金額：", bank)
```

## 6. 運算順序：括號先算

Python 的運算順序和數學很像：  
1. 括號 `()` 先算  
2. 次方 `**`  
3. 乘除 `* / // %`  
4. 加減 `+ -`


```python
print(2 + 3 * 4)
print((2 + 3) * 4)
```

### 練習：把輸入數字做一連串處理

請將使用者輸入的數字存於變數 `x` 中，接著：
1. 先加 3
2. 再乘以 5
3. 最後減掉 20


```python
x = int(input("請輸入一個數字："))

print("原本的數字是：", x)

x = (x + 3) * 5 - 20

print("計算結果是：", x)
```

## 7. 算術運算子

| 運算子 | 意義 | 例子 |
|---|---|---|
| `+` | 加法 | `3 + 2` |
| `-` | 減法 | `3 - 2` |
| `*` | 乘法 | `3 * 2` |
| `/` | 除法，結果通常是 float | `5 / 2` |
| `//` | 整除，只取商 | `5 // 2` |
| `%` | 取餘數 | `5 % 2` |
| `**` | 次方 | `2 ** 3` |


```python
a = 5
b = 2

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

### `//` 和 `%` 的生活化理解：分蘋果

有 107 顆蘋果，要平均分給 13 個人。  
- `107 // 13`：每個人可以分到幾顆  
- `107 % 13`：最後剩下幾顆


```python
apples = 107
friends = 13

print("每個人分到：", apples // friends)
print("剩下：", apples % friends)
```

### 練習：輸入蘋果數量與朋友人數

請輸入兩個數字，例如：`23 5`  
代表 23 顆蘋果分給 5 位朋友。


```python
x, y = input("請輸入蘋果數量和朋友人數，例如 23 5：").split()

x = int(x)
y = int(y)

print("蘋果數量：", x)
print("朋友人數：", y)
print("每位朋友分到：", x // y)
print("剩下：", x % y)
```

## 8. 比較運算子：得到 `True` 或 `False`

比較運算子的結果會是布林值。

| 運算子 | 意義 |
|---|---|
| `>` | 大於 |
| `<` | 小於 |
| `>=` | 大於等於 |
| `<=` | 小於等於 |
| `==` | 是否相等 |
| `!=` | 是否不相等 |


```python
a = 10
b = 12

print(a > b)
print(a < b)
print(a == b)
print(a != b)
```

### 注意：`=` 和 `==` 不一樣

- `=`：把右邊的值放進左邊的變數  
- `==`：比較左右兩邊是否相等


```python
number = 10
print(number)

number = (10 == 0)
print(number)
```

## 9. 綜合練習：比較兩個計算結果

案例一：數值 `a` 除以 20，加 8，再乘以 23。  
案例二：數值 `b` 除以 43，加 96，再乘以 6。  

請判斷哪一個比較大。


```python
a = int(input("請輸入第一個整數 a："))
b = int(input("請輸入第二個整數 b："))

a_calculation = (a / 20 + 8) * 23
b_calculation = (b / 43 + 96) * 6

print("a 的計算結果：", a_calculation)
print("b 的計算結果：", b_calculation)

print("a_calculation > b_calculation:", a_calculation > b_calculation)
print("a_calculation < b_calculation:", a_calculation < b_calculation)
print("a_calculation == b_calculation:", a_calculation == b_calculation)
```

### 挑戰 2：找零問題
輸入商品價格與付款金額，計算應該找回多少錢。  
再進一步算出可以找幾張 100 元、幾張 10 元、幾個 1 元。

### 挑戰 3：成績判斷
輸入一個分數，判斷它是否大於等於 60。  
目前先印出 `True` 或 `False` 即可。


```python
# 挑戰 1 範例解答：時間換算

total_minutes = int(input("請輸入總分鐘數："))

hours = total_minutes // 60
minutes = total_minutes % 60

print(hours, "小時", minutes, "分鐘")
```

```python
# 挑戰 2 範例解答：找零問題

price = int(input("請輸入商品價格："))
payment = int(input("請輸入付款金額："))

change = payment - price

hundreds = change // 100
change = change % 100

tens = change // 10
ones = change % 10

print("找零：", payment - price)
print("100 元：", hundreds, "張")
print("10 元：", tens, "張")
print("1 元：", ones, "個")
```

# Lesson 6：Python 輸出格式化與跳脫字元

本筆記整理 Python 常見的輸出格式控制方法：`print()`、跳脫字元、舊式 `%` 格式化、`.format()`、f-string，以及輸入後整理成指定輸出格式。


## 1. 字元與電腦內部表示

例如：

```text
'a' -> 1
'b' -> 2
'1' -> 1000
'2' -> 1001
```

前面的部分是人類可讀的字元，後面的部分可以想成是電腦用來辨識資料的編碼。


## 2. `print()`、換行與 `end`

`print()` 預設會在最後加上一個換行。可以用 `end=''` 取消自動換行。


````python
print('\n')  # 印出一個換行字元
print('h', end='')  # end='' 表示結尾不自動換行
````

## 4. 常見跳脫字元：Tab 與換行

- `	`：Tab，常用來對齊欄位。
- `
`：換行。


````python
print('Day	Frequency	Response\n0	14	0.09983')
````

## 5. 舊式 `%` 格式化

常見格式：

- `%d`：整數
- `%s`：字串
- `%f`：浮點數


````python
temp = 20
action = '走路'

print('今天天氣 %d 度' % temp)
print('要去%s' % action)
````

### 欄位寬度與小數位數

`%7.3f` 代表：

- 總寬度至少 7 格
- 小數點後保留 3 位
- 格子不足時，左邊補空白


````python
num = 5

print('%7.3f' % num)
print(num)
````

### 指定寬度輸出

例如希望輸出寬度固定為 4：

```text
12  -> __12
100 -> _100
```

這裡的 `_` 代表空白。


````python
num = 12

print('%4.0f' % num)
````

## 6. `.format()` 格式化

`.format()` 可以把變數依序放進 `{}` 中。


````python
temp = 20
action = '走路'

print('今天天氣 {} 度'.format(temp))
print('要去{}'.format(action))
````

### 使用位置編號

`{0}`、`{1}`、`{2}` 可以指定 `.format()` 中第幾個資料要放到哪裡。


````python
print('{0}{1}{0}{2}{0}{1}'.format('A', 'B', 'C'))
````

## 7. f-string 格式化

f-string 是目前最常用、也最直覺的格式化方式。只要在字串前加上 `f`，就可以直接在 `{}` 中放入變數。


````python
temp = 20
action = '走路'

print(f'今天天氣 {temp} 度')
print(f'要去{action}')
````

## 8. 浮點數保留兩位小數

`%.2f` 代表小數點後保留 2 位。


````python
num = 3.49732

print('%.2f' % num)
````

## 9. 輸入後整理成指定格式

題目需求：輸入名字和分數，輸出成：

```text
Amy: 90
```


````python
name, score = input().split()

print('%s: %s' % (name, score))
````

## 10. 印出反斜線 `\`

因為 `\` 是跳脫字元的開頭，所以如果要印出真正的反斜線，需要寫成 `\\`。


````python
print('\\')
````

## 11. 本課重點整理

1. `print()` 預設會自動換行。
2. `end=''` 可以取消 `print()` 的自動換行。
3. `	` 是 Tab，`
` 是換行。
4. `%d`、`%s`、`%f` 是舊式格式化。
5. `.format()` 可以用 `{}` 放入資料。
6. f-string 可以直接把變數寫在 `{}` 裡。
7. `%.2f` 可以控制浮點數保留兩位小數。
8. 要印出反斜線 `\`，程式碼要寫成 `\\`。

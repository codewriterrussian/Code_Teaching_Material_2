# Lesson 4：運算子 Operators

本課目標：認識 Python 中常見的運算子，包含算術運算、比較運算、邏輯運算，以及字串與串列的 `+`、`*` 用法。

> 學習重點：運算子不只看「符號」，也要看「資料型態」。同一個符號遇到不同型態時，意思可能不同。


## Section 1. 運算子的三個觀察方向

看到一個運算式時，可以先問三件事：

1. **動作**：使用的是哪一種運算？例如 `+`, `-`, `*`, `/`, `and`, `or`。
2. **規則**：運算順序是什麼？例如先乘除後加減、括號優先、`and` 比 `or` 優先。
3. **對象**：運算的資料型態是什麼？例如整數、字串、布林值、串列。


## Section 2. 布林值與 `or`

`or` 是邏輯運算子，只要其中一邊是 `True`，結果就是 `True`；只有兩邊都是 `False` 時，結果才是 `False`。


````python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

### `or` 真值表

| 第一個布林值 | 第二個布林值 | 結果 |
|---|---|---|
| `True` | `True` | `True` |
| `True` | `False` | `True` |
| `False` | `True` | `True` |
| `False` | `False` | `False` |


````python
bool_1 = True
bool_2 = False
bool_3 = bool_1 or bool_2

print("First:", bool_1, "Second:", bool_2, "Third:", bool_3)
````

## Section 3. 布林值與 `and`

`and` 也是邏輯運算子，只有兩邊都是 `True` 時，結果才是 `True`；只要有一邊是 `False`，結果就是 `False`。


````python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
````

### `and` 真值表

| 第一個布林值 | 第二個布林值 | 結果 |
|---|---|---|
| `True` | `True` | `True` |
| `True` | `False` | `False` |
| `False` | `True` | `False` |
| `False` | `False` | `False` |


## Section 4. 邏輯運算順序

在 Python 中，邏輯運算的優先順序通常是：

1. 括號 `()`
2. 比較運算，例如 `==`, `>=`, `<`
3. `and`
4. `or`

所以 `and` 會比 `or` 先運算。當運算式很長時，建議加上括號，讓意思更清楚。


````python
print(True or False and True)
# 等同於：True or (False and True)
# -> True or False
# -> True

print(True or False or False)
# -> True or False or False
# -> True

print(True and False and True or False)
# -> ((True and False) and True) or False
# -> (False and True) or False
# -> False or False
# -> False
````

````python
print(True or False and True)  # True
print(False and False or True)  # True

print(True or False and True and False or True and False or True and True)  # True
# True or (False and True and False) or (True and False) or (True and True)
# -> True or False or False or True
# -> True

print(1 == '1' or 1 >= 1 and False or True)  # True
# 1 == '1' -> False，因為整數 1 和字串 '1' 型態不同
# 1 >= 1 -> True
# False or (True and False) or True
# -> False or False or True
# -> True
````

## Section 5. 字串的 `+` 與 `*`

字串可以使用 `+` 做串接，也可以使用 `*` 重複輸出。


````python
s1 = "hello"
s2 = "hi"

print(s1 + " " + s2)
````

### 練習：印出簡單金字塔

目標輸出：

```text
*
**
***
****
```

想法：讓同一個字元重複不同次數。


````python
token = input("請輸入一個字串：")

print(token * 1)
print(token * 2)
print(token * 3)
print(token * 4)
````

也可以把換行字元 `
` 放進字串中，一次組合完整輸出。


````python
token = input("請輸入一個字串：")

result = token * 1 + 
" + token * 2 + "
" + token * 3 + "
" + token * 4 "
print(result)
````

## Section 6. 串列的 `+` 與 `*`

串列使用 `+` 時，意思是把兩個串列接在一起。串列使用 `*` 時，意思是把串列內容重複。


````python
l1 = [123]
l2 = [456]

print(l1 + l2)
````

````python
l_l_1 = [[123]]
l_l_2 = [[456]]

print(l_l_1 + l_l_2)
````

````python
l3 = [123, "456"]
l4 = [456, True]

print(l3 + l4)
````

````python
L1 = [123, "456", [789, "101112"]]
L2 = [456, True, [131415, False]]

print(L1 + L2)
print(L1 * 2)
````

## Section 7. 輸入與型態轉換

`input()` 讀進來的資料預設都是字串。如果要做數字運算，通常需要使用 `int()` 或 `float()` 轉型。


### 範例 1：計算總價

輸入單價與數量，輸出總價。


````python
price = int(input("請輸入單價："))
count = int(input("請輸入數量："))

print("總價 =", price * count)
````

### 範例 2：判斷是否及格

比較運算的結果會是布林值 `True` 或 `False`。


````python
score = int(input("請輸入分數："))

print(score >= 60)
````

### 範例 3：判斷是否可以搭乘設施

如果需要同時滿足兩個條件，可以使用 `and`。


````python
height = int(input("請輸入身高："))
age = int(input("請輸入年齡："))

print(height >= 120 and age >= 10)
````

### 範例 4：重複輸出字串

字串乘上整數，可以得到重複後的字串。


````python
word = input("請輸入一個字：")
count = int(input("請輸入重複次數："))

print(word * count)
````

## Section 8. 一行輸入多個資料

當題目要求「同一行輸入多個資料」時，可以使用 `.split()`。


````python
price, count = input("請輸入價格和數量（以空格分隔）：").split()

total_price = int(price) * int(count)
print("總價格為：", total_price)
````

````python
height, age = input("請輸入身高和年齡（以空格分隔）：").split()
height = int(height)
age = int(age)

print(height >= 120 and age >= 10)
````

## Section 9. 常見錯誤提醒

1. `1` 和 `'1'` 不一樣：前者是整數，後者是字串。
2. 字串可以和字串相加，但字串不能直接和整數相加。
3. `input()` 得到的是字串，要做數字運算前要先轉型。
4. `and` 會比 `or` 先運算；運算式太長時建議加括號。
5. 串列的 `+` 是合併串列，不是數學加法。

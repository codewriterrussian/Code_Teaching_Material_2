## 2. 布林值 `bool`

布林值只有兩種結果：

- `True`：成立、是真的
- `False`：不成立、是假的

常見資料型態轉成布林值時，可以先用下面的規則理解：

| 資料型態 | 會變成 `False` 的情況 | 其他情況 |
|---|---|---|
| 字串 `str` | 空字串 `''` | `True` |
| 整數 / 浮點數 `int` / `float` | `0` 或 `0.0` | `True` |
| 串列 `list` | 空串列 `[]` | `True` |


### 2.1 字串轉成布林值

空字串 `''` 代表「裡面沒有任何文字」，所以會變成 `False`。


```python
name = "Tom"
empty_name = ""  # 沒有任何字元

print(name)
print(empty_name)

print(bool(name))
print(bool(empty_name))  # 問：字串裡面有沒有東西？
```

### 2.2 數字轉成布林值

在 Python 裡：

- `0` 會變成 `False`
- 不是 `0` 的數字會變成 `True`


```python
num = -1
zero = 0

print(num, zero)
print(bool(num), bool(zero))
```

### 2.3 串列轉成布林值

空串列 `[]` 代表「沒有元素」，所以會變成 `False`。


```python
items = ["hi", 2, 3]
empty_items = []

print(items, empty_items)
print(bool(items), bool(empty_items))
```

## 3. 輸入、型別轉換與計算

`input()` 讀進來的資料預設是字串。

如果要做數學運算，要先用 `int()` 轉成整數。


```python
num = int(input("請輸入數量："))

total = 50 * num
print("總數是：", total)
```

## 4. 除法、商數與餘數

| 運算子 | 意思 | 範例 |
|---|---|---|
| `/` | 一般除法，結果會是浮點數 | `23 / 5` |
| `//` | 取商數 | `23 // 5` |
| `%` | 取餘數 | `23 % 5` |

例如 `23` 個人，每 `5` 人一組：

- 可以分成 `4` 組
- 剩下 `3` 人


```python
print(23 // 5)  # 商數：4
print(23 % 5)   # 餘數：3
```

## 5. 練習：分組問題

### 題目 1

有 `N` 個人，每 `K` 個人一組，且 `N > K`。

請輸出：

1. 最多可以分成幾組
2. 剩下多少人

### 題目 2

有 `N²` 個人，每 `K²` 個人一組，且 `N > K`。

請輸出：

1. 最多可以分成幾組
2. 剩下多少人


```python
N, K = input("請輸入 N 和 K，並用空格分開（N > K）：").split()

N = int(N)
K = int(K)

N_square = N ** 2
K_square = K ** 2

print("N:", N, "K:", K)

print("Without Square")
print("Group:", N // K, "Remain:", N % K)

print("Square")
print("Group:", N_square // K_square, "Remain:", N_square % K_square)
```

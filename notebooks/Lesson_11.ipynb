## 1. 追蹤 For 迴圈中的變數

````python
score = 3
bonus = 1

for round_num in range(1, 5):
    score += round_num
    bonus *= 2

    if score % 2 == 0:
        score += bonus

    print(round_num, score, bonus)
````

### 執行結果追蹤

| `round_num` | `score` | `bonus` |
|---:|---:|---:|
| 1 | 6 | 2 |
| 2 | 12 | 4 |
| 3 | 15 | 8 |
| 4 | 19 | 16 |

## 2. Continue：跳過本次迴圈

````python
total = 0

for number in range(2, 11):
    if number % 3 == 0:
        total += 1
        continue

    total += number
    print(number, total)

print("final:", total)
````

當數字可以被 `3` 整除時，程式只會把 `1` 加到 `total`，然後使用 `continue` 跳到下一次迴圈。

最後的 `total` 是 `39`。

## 3. Break：符合條件時停止迴圈

````python
value = 1

for i in range(1, 8):
    value *= 2

    if value > 20:
        break

    print(i, value)

print("end:", value)
````

當 `value` 變成 `32` 時，條件 `value > 20` 成立，因此迴圈停止。

迴圈中的輸出為 `2、4、8、16`，最後印出的 `value` 是 `32`。

## 4. 反向 Range 與累加

````python
result = 0

for number in range(15, 2, -3):
    if number % 2 == 0:
        result -= number
    else:
        result += number

    print(result)
````

`range(15, 2, -3)` 會依序產生：

```text
15, 12, 9, 6, 3
```

奇數加到 `result`，偶數則從 `result` 中減去。

## 5. 使用索引處理串列

````python
values = [4, 7, 2, 9]
result = 0

for i in range(len(values)):
    if i % 2 == 0:
        result += values[i]
    else:
        result -= values[i]

    print(i, result)
````

### 計算相鄰數字變大的次數

````python
numbers = [3, 7, 5, 8, 8, 12, 4]
count = 0

for i in range(len(numbers) - 1):
    if numbers[i + 1] > numbers[i]:
        count += 1

print(count)
````

### 找出第一個下降的位置

````python
numbers = [2, 5, 9, 11, 8, 13]

for i in range(1, len(numbers)):
    if numbers[i] < numbers[i - 1]:
        print(i)
        break
````

## 6. 綜合使用 Continue 與 Break

````python
points = 0

for number in range(1, 13):
    if number % 4 == 0:
        continue

    if number % 3 == 0:
        points += 5
    else:
        points += 1

    if points >= 15:
        break

    print(number, points)

print("score:", points)
````

執行順序：

1. 可以被 `4` 整除的數字會被跳過。
2. 可以被 `3` 整除的數字加 `5` 分，其他數字加 `1` 分。
3. 當分數達到 `15` 分以上時，迴圈停止。

## 7. While 迴圈中的數值變化

````python
number = 3

while number < 25:
    print(number)

    if number % 2 == 0:
        number += 5
    else:
        number *= 2
````

### 使用 Continue 略過指定數字

````python
number = 0

while number < 8:
    if number == 3:
        number += 1
        continue

    print(number)
    number += 1
````

## 8. While 實作：角色生命值

````python
hp = 31
turn = 0

while hp > 0:
    turn += 1

    if turn % 3 == 0:
        hp += 4
    else:
        hp -= 9

    print(turn, hp)
````

每三回合恢復 `4` 點生命值，其他回合損失 `9` 點生命值。當 `hp <= 0` 時，迴圈結束。

## 9. While True：持續接收輸入

````python
total = 0
valid_count = 0

while True:
    number = int(input("請輸入整數，輸入 0 結束："))

    if number == 0:
        break

    if number < 0:
        continue

    total += number
    valid_count += 1

print("總和：", total)
print("有效數量：", valid_count)
````

規則：

- 輸入 `0`：結束迴圈
- 輸入負數：不計入結果
- 輸入正數：加入總和，並增加有效數量

## 10. 索引綜合練習

### 奇偶索引交替加減

````python
numbers = [10, 3, 6, 2, 5]
total = 0

for i in range(len(numbers)):
    if i % 2 == 0:
        total += numbers[i]
    else:
        total -= numbers[i]

print(total)
````

### 計算相鄰元素的差

````python
numbers = [4, 9, 6, 10]
differences = [0] * (len(numbers) - 1)

for i in range(len(numbers) - 1):
    differences[i] = numbers[i + 1] - numbers[i]

print(differences)
````

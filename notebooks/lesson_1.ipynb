## 二、使用 `print()` 輸出文字

`print()` 可以請電腦把括號裡的內容顯示出來。

可以把它想成：請電腦「說出一句話」。


````python
print("Hi")
````

### 單引號與雙引號

在 Python 中，文字可以放在雙引號 `" "` 裡，也可以放在單引號 `' '` 裡。

只要前後使用一致，兩種寫法通常都可以。


````python
print('hi')
````

## 三、文字數字與真正數字的差別

下面兩個輸出看起來很像，但對電腦來說意義不同。

- `"2345"` 是文字，也就是字串。
- `2345` 是數字，可以做數學運算。

生活例子：

- 電話號碼、學生證號碼通常比較像文字，因為我們不會拿它們做加減乘除。
- 分數、金額、溫度則是數字，因為可以拿來計算。


````python
print("2345")
````

````python
print(2345)
````

## 四、補充觀念：數學式與程式

在數學課裡，我們可能會看到：

```text
x + y = 1
x + 2y = 5
```

但在 Python 裡，等號 `=` 通常不是「數學上的相等」，而是「把右邊的資料放進左邊的變數」。

例如：

```python
a = 10
```

意思是把 `10` 放進變數 `a` 裡。


## 五、使用變數儲存資料

變數可以想成一個盒子，用來存放資料。

下面的例子中：

- `a` 裡面放的是文字 `"2345"`。
- `b` 裡面放的是數字 `2345`。

雖然印出來看起來相似，但它們的資料型態不同。


````python
a = "2345"
b = 2345
print(a)
print(b)
````

## 六、布林值簡介：判斷是否相等

`!=` 的意思是「不等於」。

下面這一行會判斷 `1` 是否不等於 `0`。因為這是真的，所以結果會是 `True`。


````python
1 != 0
````

## 七、使用 `type()` 檢查資料型態

`type()` 可以讓我們知道某個資料在 Python 裡屬於哪一種型態。

這在除錯時很重要，因為很多錯誤其實是「資料型態不對」造成的。


````python
type(a) # -> str = string
````

`str` 是 string 的縮寫，意思是「字串」，也就是文字資料。


````python
type(b) # -> int = integer
````

`int` 是 integer 的縮寫，意思是「整數」。

整數可以用來做數學計算，例如加、減、乘、除。


## 八、電腦與人類讀資料的方式

可以用下面的方式幫助學生理解：

- 電腦做計算時，主要讀的是數字。
- 人類閱讀時，主要讀的是文字。
- 在程式中，文字通常使用字串 `str` 表示。

原本課堂筆記：

```text
Computer -> coding -> Read (Numbers)
Human -> Words -> String for Reading
```


## 九、建立文字變數

我們可以把一段文字存進變數中。


````python
text1 = "Hi"
type(text1)
````

這裡 `text1` 存的是 `"Hi"`，所以它的型態是 `str`。


## 十、數字相加

如果兩邊都是數字，`+` 會執行數學加法。


````python
print(123+455)
````

## 十一、字串相加

如果兩邊都是字串，`+` 會把文字接在一起，這叫做字串串接。


````python
print("hi" + " there")
````

這裡不是數學加法，而是把 `"hi"` 和 `" there"` 接在一起，變成一句完整的話。


## 十二、輸出完整句子

我們可以直接輸出一整句文字。


````python
print("Today is 24 celius degrees")
````

## 十三、使用變數組合句子

如果溫度會改變，就可以把溫度放在變數中。

不過，因為 `temp` 是數字，不能直接和字串相加，所以要先用 `str(temp)` 把它轉成文字。


````python
temp = 28
print("Today is " + str(temp) + " celius degrees")

# print("Today is " + "celius degrees")
````

## 十四、講師提醒

這一節課最容易混淆的地方是：

1. `"2345"` 和 `2345` 看起來一樣，但型態不同。
2. `+` 在數字中是加法，在字串中是連接文字。
3. 數字和文字不能直接用 `+` 接在一起。
4. 如果要把數字放進句子裡，通常要先使用 `str()` 轉換。

可以提醒學生：電腦很嚴格，看到文字就是文字，看到數字就是數字。即使畫面上看起來一樣，對電腦來說也可能完全不同。

## 十六、課程總結

今天的重點是讓學生理解 Python 最基本的輸出與資料型態。

簡單來說：

- `print()`：把內容顯示出來。

- `int`：整數資料。

# Lesson 01：`print()` 與 `input()` 基礎

> 這堂課的重點：學會把文字與數字印出來，理解 `print()` 的分隔與換行設定，並使用 `input()` 讓使用者輸入資料。


## Section I. 第一個輸出：Hello World

`print()` 可以把括號裡面的內容顯示在畫面上。文字資料需要用引號包起來。


````python
# 印出一段文字
print('Hello World')

# 也可以印出中文
print('早安')
````

## Section II. 文字和數字的差別

在 Python 裡，`3` 是數字，`'3'` 是文字。雖然看起來都像 3，但資料型態不同。


````python
# 文字 hi
print('hi')

# 數字 3
print(3)

# 文字 3
print('3')
````

## Section III. 在 `print()` 裡做計算

`print()` 裡面可以直接放算式，Python 會先算完，再把結果印出來。


````python
# 加法
print(123 + 456)

# 乘法
print(10 * 5)

# 也可以把文字和算式一起印出
print(124, '可以', '目前', 23 * 11)
````

## Section IV. 一次印出多個資料

`print()` 可以一次放很多個資料，中間用逗號 `,` 分開。預設輸出時，每個資料中間會有一個空格。


````python
print(123, '早安', 'hi')
print(1, 2, 3, 4, 5)
````

## Section V. `sep`：控制資料之間的分隔符號

`sep` 是 separator 的縮寫，意思是「分隔符號」。它可以控制多個輸出資料中間要放什麼。


````python
# 預設是空格分隔
print(1, 2, 3, 4, 5)

# 明確指定用空格分隔
print(1, 2, 3, 4, 5, sep=' ')

# 用逗號分隔
print(1, 2, 3, 4, 5, sep=',')

# 完全不要分隔
print(1, 2, 3, 4, 5, sep='')
````

### 練習：城市路線輸出

請觀察 `sep=' -> '` 和 `sep=' => '` 的效果。


````python
print('台北', '台中', '高雄', sep=' -> ')
print('台北', '東京', '首爾', '紐約', '柏林', '華沙', '莫斯科', sep=' => ')
````

## Section VI. `end`：控制輸出結尾

`print()` 預設每次印完後都會自動換行。這個預設結尾其實是 `end='\n'`。

如果把 `end` 改成空字串，就不會自動換行。


````python
# 預設會自動換行
print(123)
print(456)

# 不自動換行
print(123, end=' ')
print(456)

print(123, 456)
````

### `
` 是什麼？

`
` 代表換行。也就是讓輸出的文字跳到下一行。


````python
print(123, end=' ')
print(598 ^ 999)
print(70)
````

## Section VII. `input()`：讓使用者輸入資料

`input()` 會讓程式暫停，等待使用者輸入內容。使用者按下 Enter 後，程式才會繼續執行。


````python
# 讓使用者輸入，並直接印出輸入內容
print(input())
````

### 加上提示文字

可以在 `input()` 的括號中放提示文字，讓使用者知道要輸入什麼。


````python
print('你好', input('請輸入你的名字：'))
````

## Section VIII. 情境練習：自助點餐機

自助點餐機問你「請輸入取餐號碼」，你輸入後，機器才能顯示你的訂單。


````python
number = input('請輸入你的號碼：')

print('你的號碼是 =', number, '你的餐點是', '牛肉麵')
print(number, '-> 是你的號碼', '你的餐點是', '牛肉麵', sep='  ')
````

## Section IX. 常見錯誤提醒

### 1. 不要把 `input` 當成資料本身

錯誤觀念：

```python
data1 = input
```

這只是把 `input` 這個函式本身存起來，並沒有真的讓使用者輸入。

正確寫法：

```python
data1 = input('請輸入資料：')
```


````python
data1 = input('請輸入第一筆資料：')
data2 = input('請輸入第二筆資料：')

print('第一筆資料：', data1)
print('第二筆資料：', data2)
````

## 1. 輸入與輸出：Hello 題型

**題目目標**：讀入一個字串，輸出 `hello, 字串`。

**重點觀念**：
- `input()` 會讀取使用者輸入，結果是「字串」。
- `print()` 會把內容顯示在畫面上。
- 字串可以用 `+` 連接。


````python
name = input("請輸入名字或單字：")
print("hello, " + name)
````

**練習**：把輸出改成 `Hi, xxx!`，例如輸入 `Tom`，輸出 `Hi, Tom!`。

## 2. 讀入兩個數字並相加

**題目目標**：讀入兩個整數，輸出它們的和。

**重點觀念**：
- `input().split()` 可以把一行輸入切成多個字串。
- `int()` 可以把數字字串轉成整數。
- 從 `input()` 讀進來的資料預設都是字串，所以做加法前要轉成 `int`。


````python
a_text, b_text = input("請輸入兩個整數，中間用空白隔開：").split()

a = int(a_text)
b = int(b_text)

print(a + b)
````

**範例輸入**：`3 5`  
**範例輸出**：`8`

## 3. 運勢判斷：月份與日期

**題目目標**：輸入月與日，計算：

```text
S = (月份 * 2 + 日期) % 3
```

然後依照結果輸出：
- `S == 0`：普通
- `S == 1`：吉
- `S == 2`：大吉

**重點觀念**：
- `%` 是取餘數。
- `if / elif / else` 可以依照不同條件執行不同程式。


````python
month_text, day_text = input("請輸入月份與日期，例如 8 27：").split()

month = int(month_text)
day = int(day_text)

s = (month * 2 + day) % 3

if s == 0:
    print("普通")
elif s == 1:
    print("吉")
else:
    print("大吉")
````

**學生提醒**：這題常見錯誤是把 `S == 2` 寫成其他結果。根據題目，`2` 對應的是 `大吉`。

## 4. 閏年判斷

**題目目標**：輸入一個西元年份，判斷是不是閏年。

**判斷規則**：
- 能被 400 整除：一定是閏年。
- 能被 4 整除，而且不能被 100 整除：是閏年。
- 其他情況：不是閏年。

**重點觀念**：
- `==` 是判斷是否相等。
- `and` 表示「而且」。
- `or` 表示「或者」。


````python
year = int(input("請輸入年份："))

is_leap_year = (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)

if is_leap_year:
    print("閏年")
else:
    print("平年")
````

**可以測試的年份**：
- `2000`：閏年，因為能被 400 整除。
- `1900`：平年，雖然能被 4 整除，但也能被 100 整除，而且不能被 400 整除。
- `2024`：閏年，因為能被 4 整除，而且不能被 100 整除。

## 5. 絕對值：把負數變正數

**題目目標**：輸入一個整數，輸出它的絕對值。

**重點觀念**：
- `abs(x)` 會回傳 `x` 的絕對值。
- 負數會變正數，正數不變，`0` 仍然是 `0`。


````python
number = int(input("請輸入一個整數："))
print(abs(number))
````

### 另一種想法：用字串處理負號

> 這題的重點不是練習絕對值的最簡寫法，而是複習字串處理。  
> 我們先把輸入當成字串，再用 `split("-")` 把負號切開，最後取出數字部分。  
> 如果只是單純要求絕對值，通常可以直接使用 `abs()`；但這題想複習的是 `split()` 的用法。

````python
text = input("請輸入一個整數：")
parts = text.split("-")

# 如果輸入是 -25，split 後會得到 ['', '25']，最後一項是 '25'
# 如果輸入是 25，split 後會得到 ['25']，最後一項也是 '25'
absolute_text = parts[-1]

print(absolute_text)
````

## 6. List 與基本運算子

**重點觀念**：
- List 可以存放多個資料。
- `list[-1]` 代表最後一個元素。
- `+=`、`-=`、`*=`、`//=` 是更新變數的簡寫。


````python
items = ["2", 6, 8, 10, "a"]
print(items[-1])  # 取出最後一個元素
````

````python
x = 5
x += 1  # 等同於 x = x + 1
print(x)

x *= 2  # 等同於 x = x * 2
print(x)

x //= 3  # 整數除法，只取商
print(x)
````

### 建立一個每次乘以 2 的列表

輸入一個數字，建立：`[原數字, 2倍, 4倍, 8倍, 16倍, 32倍]`。


````python
start = int(input("請輸入起始數字："))

numbers = [start]
current = start

for i in range(5):
    current *= 2
    numbers.append(current)

print(numbers)
````

## 7. 比較運算與字串比較

**重點觀念**：
- `==` 判斷兩個值是否一樣。
- 字串比較不是看數字大小，而是依照字元順序比較。
- 大小寫不同，字串也不同，例如 `hello` 和 `Hello` 不相等。


````python
print(10 == 10)       # True
print(10 == "10")     # False，整數 10 和字串 "10" 不同
print("hello" == "Hello")  # False，大小寫不同
````

````python
print("10" < "2")

# 注意：這是字串比較，不是數字比較。
# 字串會從第一個字元開始比，所以 "1" 會排在 "2" 前面。
````

````python
a = "20,15.00,Hi"
b = "20, 15.00, Hi"

print(a == b)

# b 裡面多了空白，所以兩個字串不相等。
````

## 9. 十進位與二進位

**重點觀念**：
- 十進位使用 0 到 9。
- 二進位只使用 0 和 1。
- `bin(number)` 可以把整數轉成二進位字串。
- 不斷 `% 2` 可以取得二進位的最後一位，不斷 `// 2` 可以把數字往右移一位。


````python
number = int(input("請輸入一個十進位整數："))
print(bin(number))
````

### 手動觀察 `% 2` 與 `// 2`

下面的程式會示範：每次除以 2，餘數就是二進位的一個位數。


````python
number = int(input("請輸入一個十進位整數："))

while number > 0:
    remainder = number % 2
    print(f"目前數字：{number}，除以 2 的餘數：{remainder}")
    number //= 2
````

## 10. 字串格式化

**題目目標**：把變數放進句子裡。

**重點觀念**：
- 推薦初學者優先使用 f-string，因為最直覺。
- f-string 寫法：`f"文字 {變數} 文字"`。


````python
name = "John"
score = 95

print("hi " + name + "!")
print(f"hi {name}!")
print(f"{name} 的分數是 {score} 分")
````

### 其他常見格式化方式

這些寫法也會在舊程式中看到，但初學時可以先熟悉 f-string。


````python
name = "John"

print("hi %s!" % name)
print("hi {}!".format(name))
print("%d, %s" % (20, "20"))
````

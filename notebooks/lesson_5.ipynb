## 1. 時間換算：秒數轉成時、分、秒

題目：輸入一個總秒數，轉換成：

```text
小時 分鐘 秒
```

例如：

```text
60 秒 = 0 小時 1 分 0 秒
2323 秒 = 0 小時 38 分 43 秒
```

思考方式：

1. `1 分鐘 = 60 秒`
2. `1 小時 = 60 分鐘 = 3600 秒`
3. 用 `//` 取商數，用 `%` 取剩下的秒數


````python
second_total = int(input("請輸入總秒數："))

hours = second_total // 3600
minutes = second_total // 60 % 60
seconds = second_total % 60

print("Second total:", second_total)
print("hours:", hours, "minutes:", minutes, "seconds:", seconds)
````

## 2. 時間換算：小時轉成天與小時

例如 `56` 小時：

- `56 // 24 = 2` 天
- `56 % 24 = 8` 小時


````python
hour_total = 56

days = hour_total // 24
hours = hour_total % 24

print(days, hours)
````

## 3. `=`、`==`、`!=` 的差別

| 符號 | 意思 | 範例 |
|---|---|---|
| `=` | 賦值，把右邊的值放進左邊變數 | `name = "John"` |
| `==` | 判斷是否相等 | `name1 == name2` |
| `!=` | 判斷是否不相等 | `name1 != name2` |

注意：`=` 不是「是否相等」，而是「把值放進變數」。


### 3.1 賦值範例


````python
name1 = "John"
name2 = "Mary"

print(name1, name2)

name1 = name2  # 把 name2 的值放進 name1

print(name1)
````

### 3.2 不相等比較


````python
name1 = "John"
name2 = "Mary"

print(name1, name2)
print(name1 != name2)
````

## 5. 邏輯運算：`and`

`and` 的意思是「而且」。

全部條件都要是 `True`，結果才會是 `True`。


````python
T = True
F = False

print(T and T and T)
print(T and F and T)
print(T and F and F)
print(F and T and T)
print(F and T and F)
print(F and F and F)
print(F and F and T)
print(T and T and F)
````

## 6. 邏輯運算：`or`

`or` 的意思是「或者」。

只要其中一個條件是 `True`，結果就是 `True`。


````python
print(T or T or T)
print(T or F or T)
print(T or F or F)
print(F or T or T)
print(F or T or F)
print(F or F or F)
print(F or F or T)
print(T or T or F)
````

## 7. 練習：搭乘限制

題目：

身高超過 `120` 公分，而且年齡超過 `10` 歲，才可以搭乘。

需要同時符合：

1. `height > 120`
2. `age > 10`


````python
height, age = input("請輸入身高和年齡，並用空格分開：").split()

height = int(height)
age = int(age)

can_ride = height > 120 and age > 10

print(can_ride)
````

## 8. 邏輯運算順序

在 Python 中，邏輯運算的優先順序通常是：

1. 比較運算，例如 `==`、`>=`
2. `and`
3. `or`

如果不確定順序，可以用括號 `()` 讓程式更清楚。


````python
print(True or False and True)
# 等同於：True or (False and True)

print(False and False or True)
# 等同於：(False and False) or True

print(True or False and True and False or True and False or True and True)

print(1 == "1" or 1 >= 1 and False or True)
````

## 1. 運算順序與次方

Python 會先處理括號，再處理次方，接著處理乘除，最後才是加減。  
`**` 是次方運算。

```python
print(2 ** 3 * 2)
print(2 * 3 ** 2)
print((2 + 3) ** 2)
```

## 2. 用 `%` 和 `//` 拆三位數

假設輸入 `728`：

- `num % 10` 可以取得個位數 `8`
- `num // 10` 可以去掉個位數，得到 `72`
- 再重複同樣方法，就可以取得十位數與百位數

```python
# 例：728 => 7 2 8

num = int(input("請輸入三位數："))

ones = num % 10
remaining = num // 10

tens = remaining % 10
hundreds = remaining // 10

print(hundreds, tens, ones)
```

## 3. 變數重新指定

變數可以被重新指定。每次指定後，原本的值會被新的值取代。

```python
x = 35

x = x // 4
print(x)

x = x * 10 + 3
print(x)

x = x % 7
print(x)
```

## 4. 字串格式化

舊式格式化常見符號：

- `%d`：整數
- `%s`：字串
- `%f`：浮點數

```python
print("%d" % 20)
print("%s" % 20)
print("%f" % 5)
```

`.format()` 可以把資料放進 `{}` 裡，也可以用位置編號調整輸出順序。

```python
print("{} {}".format(20, "Hi"))
print("{1} {0}".format("A", "B"))
```

## 5. 基本 `if` 條件判斷

`if` 下面縮排的程式，只有在條件成立時才會執行。  
沒有縮排的程式，不屬於 `if`，所以一定會執行。

```python
x = -3

if x > 0:
    print("A")

print("B")
```

## 6. `if / elif / else`

當條件有多種可能時，可以使用：

- `if`：第一個條件
- `elif`：其他條件
- `else`：以上條件都不成立時執行

```python
user_input = 3

if user_input == 1:
    print(1)
elif user_input == 2:
    print(2)
elif user_input == 3:
    print(3)
else:
    print("不是 1、2、3")
```

## 7. `elif` 只會選一個分支

在 `if / elif / else` 結構中，只要前面的條件成立，後面的 `elif` 就不會再被檢查。

```python
x = 5

if x > 0:
    print("A")
elif x < 10:
    print("B")
else:
    print("C")
```

## 8. 兩個獨立的 `if`

如果寫成兩個獨立的 `if`，Python 會分別檢查兩組條件。

```python
x = 20

if x > 0:
    print("A")
else:
    print("C")

if x < 10:
    print("B")
```

## 11. 巢狀條件判斷：成績等第

巢狀條件判斷就是在 `if` 裡面再寫一層 `if`。  
這個例子會依照分數輸出 A、B、C、F 或 X。

```python
score = int(input("請輸入分數："))

if score >= 80:
    if score > 90:
        print("A+")
    elif score > 85:
        print("A")
    else:
        print("A-")
elif score >= 70:
    if score > 77:
        print("B+")
    elif score > 73:
        print("B")
    else:
        print("B-")
elif score >= 60:
    if score > 67:
        print("C+")
    elif score > 63:
        print("C")
    else:
        print("C-")
elif 0 < score < 60:
    print("F")
else:
    print("X")
```

## 14. 用字串索引拆解數字

如果把輸入保留成字串，就可以用索引直接取得每一位數。  
例如：`1728` 的索引如下：

- `num[0]` 是 `'1'`
- `num[1]` 是 `'7'`
- `num[2]` 是 `'2'`
- `num[3]` 是 `'8'`

```python
# 1728 => 1 7 2 8
# 728  => 7 2 8

num = input("請輸入 1 到 4 位數：")

if len(num) == 4:
    print(int(num[0]), int(num[1]), int(num[2]), int(num[3]))
elif len(num) == 3:
    print(int(num[0]), int(num[1]), int(num[2]))
elif len(num) == 2:
    print(int(num[0]), int(num[1]))
elif len(num) == 1:
    print(int(num[0]))
else:
    print("請輸入 1 到 4 位數")
```

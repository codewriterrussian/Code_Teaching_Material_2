## 先學會怎麼填答案：`def` 函式作答方式

這份練習不是讓你直接使用 `input()`，而是讓你把答案寫在一個「函式 function」裡。

你會看到像這樣的格式：

```python
def q1_greet(name):
    # TODO: 回傳 Hello, 名字!
    return None
```

請注意：

1. `def q1_greet(name):` 這一行不要刪掉。
2. `name` 是題目給你的資料，可以直接拿來用。
3. `return None` 代表「還沒寫答案」，你要把 `None` 換成你的答案。
4. 這份練習要用 `return`，不是用 `print()`。

小範例：

```python
def example_name(name):
    return name
```

這代表：把收到的 `name` 直接交回去。

````
# 範例：這不是正式題目，只是教你怎麼填答案

def example_name(name):
    return name

print(example_name("Amy"))
print(example_name("Bob"))
````

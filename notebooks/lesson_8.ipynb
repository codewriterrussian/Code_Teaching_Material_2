## 1. `if` 與縮排

觀察哪些 `print()` 在 `if` 裡面，哪些在 `if` 外面。

````python
x = 3

if x > 0:
    print("positive")
    print("inside")

print("outside")
````

## 2. 分數判斷：pass / fail

規則：
- `score >= 60`：pass
- 其他：fail

````python
score = int(input("Enter score: "))

if score >= 60:
    print("pass")
else:
    print("fail")
````

## 3. 邏輯運算子：`and` / `or`

注意：`and` 的優先順序比 `or` 高。  
需要時可以使用括號讓條件更清楚。

````python
x = 5

if x < 3 or (x > 10 and x % 2 == 0):
    print("A")
else:
    print("B")
````

## 4. 除法、整除與取餘數

- `/`：一般除法
- `//`：整除，只取商
- `%`：取餘數

````python
print(5 / 3)
print(5 // 3)
print(5 % 3)
````

## 5. 月份判斷季節

規則：
- `3, 4, 5`：spring
- `6, 7, 8`：summer
- `9, 10, 11`：autumn
- `12, 1, 2`：winter
- 其他：invalid

````python
month = int(input("Enter month: "))

if month < 1 or month > 12:
    print("invalid")
elif 3 <= month <= 5:
    print("spring")
elif 6 <= month <= 8:
    print("summer")
elif 9 <= month <= 11:
    print("autumn")
else:
    print("winter")
````

## 6. 分數等級判斷

規則：
- `score >= 80`：good
- `60 <= score < 80`：pass
- `score < 60`：fail

````python
score = int(input("Enter score: "))

if score >= 80:
    print("good")
elif score >= 60:
    print("pass")
else:
    print("fail")
````

## 7. BMI 判斷

公式：

```python
BMI = kg / (m ** 2)
```

輸入格式：

```text
kg m
```

規則：
- 小於 `18.5`：underweight
- `18.5` 到小於 `24`：normal
- `24` 到小於 `27`：overweight
- `27` 以上：obese

````python
weight, height = input("Enter weight and height: ").split()

weight = float(weight)
height = float(height)

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("underweight")
elif bmi < 24:
    print("normal")
elif bmi < 27:
    print("overweight")
else:
    print("obese")
````

## 8. 一次輸入多個值

這一格示範：
- `split()` 拆開輸入
- `float()` / `int()` 型別轉換
- 用條件判斷檢查資料

````python
value_1, operator, value_2 = input("Enter: number operator number: ").split()

value_1 = float(value_1)
value_2 = float(value_2)

if operator == "+":
    print(value_1 + value_2)
elif operator == "-":
    print(value_1 - value_2)
elif operator == "*":
    print(value_1 * value_2)
elif operator == "/":
    print(value_1 / value_2)
else:
    print("invalid operator")
````

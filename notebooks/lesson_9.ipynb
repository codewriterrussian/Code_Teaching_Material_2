`range(3)` 會產生：

```text

```

注意：Python 通常從 `0` 開始數。

````python
print(list(range(3)))

numbers = [0, 1, 2]
print(numbers[0])
````

## 3. `for` 迴圈基本寫法

````python
for i in range(3):
    print(i)

print("done")
````

````python
for i in range(5):
    print(1)
````

## 7. `range(start, stop)`：指定開始與結束

注意：尾巴不包含 `stop`。

````python
numbers = list(range(2, 21))

print(len(numbers))
print(numbers)
````

## 8. `range(start, stop, step)`：指定間隔

````python
even_numbers = list(range(0, 12, 2))
odd_numbers = list(range(1, 12, 2))

print(even_numbers)
print(odd_numbers)
````

## 9. 走訪串列：天氣判斷

規則：
- 晴天：出門曬太陽
- 陰天：睡覺

````python
weather = ["晴天", "陰天", "晴天", "晴天", "陰天"]

for w in weather:
    if w == "晴天":
        print("出門曬太陽")
    elif w == "陰天":
        print("睡覺")
````

## 10. 串列中的值不會自動改變

在迴圈中寫 `t = t + 1`，只會改變暫時變數 `t`，不會改變原本的 `temps`。

````python
temps = [31, 27, 27, 27, 26, 27, 27]

for t in temps:
    t = t + 1
    print(t)

print(temps)
````

## 12. 1 到 n 的總和

輸入格式：

```text
n
```

目標：
- 印出奇數和
- 印出偶數和
- 印出總和

````python
n = int(input("Enter n: "))

total_odd = 0
total_even = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        total_even = total_even + i
    else:
        total_odd = total_odd + i

print("Sum of odd numbers:", total_odd)
print("Sum of even numbers:", total_even)
print("Total sum:", total_odd + total_even)
````

## 13. 計算偶數有幾個

````python
numbers = [1, 2, 3, 4, 5]
count = 0

for number in numbers:
    print("item in numbers: {}".format(number))

    if number % 2 == 0:
        count = count + 1

print("count of even numbers: {}".format(count))
````

## 14. 找出最大值

做法：
1. 先把第一個數字當成目前最大值。
2. 每次拿新的數字來比較。
3. 如果新的數字比較大，就更新最大值。

````python
target = [3, 1, 8, 2]
max_num = target[0]

for t in target:
    if t > max_num:
        max_num = t

print(max_num)
````

## 15. 印出正數

範例：
- `[-2, 0, 3, 5, -1]` → `3 5`
- `[2, 4, 6, -2, 9, 0]` → `2 4 6 9`

````python
numbers = [-2, 0, 3, 5, -1]

for number in numbers:
    if number > 0:
        print(number)
````

## 16. 簡易計算機

輸入格式：

```text
a b op
```

範例：

```text
3 5 +
```

````python
a, b, op = input("Enter: a b op: ").split()

a = int(a)
b = int(b)

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
else:
    print("Invalid operator")
````

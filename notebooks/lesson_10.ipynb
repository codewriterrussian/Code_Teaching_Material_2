## 2. 走訪串列

```python
names = ["Amy", "Ben", "Cindy"]

for name in names:
    print(name)

for _ in range(len(names)):
    print(names[_])
```

## 3. 計數與累加

```python
numbers = [4, 7, 2, 9]
lst_sum = 0

for num in numbers:
    lst_sum += num

print(lst_sum)
```

## 4. 奇數與偶數累加

```python
# 4 -> 1+3; 2+4

n = int(input())

total_odd = 0
total_even = 0

for num in range(n+1):
    if num % 2 == 0:
        total_even += num
    else:
        total_odd += num

print(total_even, total_odd)
```

## 5. 比較最大值

```python
numbers = [14, 3, 27, 9, 21]

# 14 -> 3 -> 27 -> 9 -> 21
# 初始比較數值 -> 0
# 14 > 0 => 比較數值 -> 14

num_compare = None

for num in numbers:
    if num_compare is None:
        num_compare = num

    if num_compare < num:
        num_compare = num

print(num_compare)
```

```python
numbers = [-8, -3, -12, -5]
largest = None

for number in numbers:
    if largest is None:
        largest = number
    if number > largest:
        largest = number

print(largest)
```

## 6. While 迴圈

### `for` 與 `while` 的差別

- `for`：通常用在已知範圍或有限次數的重複工作。
- `while`：只要條件成立就持續執行，直到條件不成立。

例如：體力值從 `107` 開始，每次減少 `10`，可以使用條件決定何時停止。

```python
i = 0

# 0 -> True -> 1 -> True -> 2 -> True -> 3 -> False => X
while i < 3:
    print(i)
    i += 1
```

```python
for i in range(3):
    print(i)
```

## 7. While 實作：登入流程

```python
# 再給一個條件: 帳號對密碼錯，印出 ‘密碼錯誤’， 下次就只問密碼

username = "hello"
password = "123"
login = False
ask_user_falg = False

while not login:

    if not ask_user_falg:
        username_input = input('username')

    password_input = input('password')

    if username_input == username and password_input == password:
        login = True
        print("已登入")

    elif username_input == username:
        ask_user_falg = True
        print('密碼錯誤')

    else:
        print("錯誤")
```

## 8. Break 與 Continue

### 執行流程

目標數字：`0 1 2 3 4 5 6 7`

- `break`：遇到指定條件時直接停止迴圈。
- `continue`：跳過本次迴圈，繼續下一次。

例如遇到 `3`：

- `break` → `0 1 2`
- `continue` → `0 1 2 4 5 6 7`

```python
for i in range(20):
    if i % 4 == 0:
        continue
    elif i == 10:
        break
    print(i)
```

```python
# continue, break
# 1. 產生 for loop 13 -> 124
# 2. 假如 可以被 27 整除 跳過
# 3. 遇到 101 停下來
# 13, 14, ... , 100

num = 13

while num < 125:
    if num % 27 == 0:
        num += 1
        continue
    elif num == 101:
        break

    print(num)
    num += 1
```

```python
lst = [1, 2, 3, 4, 6]
cnt = 0

for l in lst:
    if l % 2 == 0:
        cnt += 1

print(cnt)

print(len(lst))
print(lst[0],lst[1],lst[2])

# while + continue

idx = -1
cnt = 0

while idx < len(lst)-1:
    idx += 1
    if lst[idx] % 2 != 0:
        continue
    cnt += 1
print(cnt)
```

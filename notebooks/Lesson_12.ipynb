## 1. 輸入、處理與輸出

程式通常可以分成三個部分：

1. **輸入**：取得資料
2. **處理**：執行計算或判斷
3. **輸出**：顯示或回傳結果

````python
numbers = [3, 4, 5]
total = 0

for number in numbers:
    total += number

print(total)
````

## 2. 定義函式

函式的基本格式：

```python
def function_name(parameter_1, parameter_2):
    # 執行內容
    return result
```

括號內放入函式需要接收的資料，稱為**參數**。

### Print 與 Return 的差異

````python
def add_and_print(a, b):
    print(float(a) + float(b))


answer = add_and_print(3, 5)
print(answer)  # 函式沒有 return，因此 answer 是 None
````

````python
def add_num(a, b):
    return float(a) + float(b)


answer = add_num(3, 5)
print(answer)
````

- `print()`：把內容顯示在畫面上。
- `return`：把結果送回函式外，讓其他程式繼續使用。
- 函式沒有寫 `return` 時，預設回傳 `None`。

## 3. 預設參數

````python
def say_hello(text, times=3):
    for _ in range(times):
        print(text)


say_hello("hi", 10)
````

`times=3` 是預設參數。呼叫函式時沒有提供 `times`，函式就會使用預設值 `3`。

## 4. 回傳多個結果

以下函式會分別計算：

- 偶數索引位置的總和：`0、2、4...`
- 奇數索引位置的總和：`1、3、5...`

````python
def index_sum(values):
    even_index_sum = 0
    odd_index_sum = 0

    for i in range(len(values)):
        if i % 2 == 0:
            even_index_sum += values[i]
        else:
            odd_index_sum += values[i]

    return even_index_sum, odd_index_sum
````

````python
numbers = [1, 2, 4, 5]

even_sum, odd_sum = index_sum(numbers)
print("偶數索引總和：", even_sum)
print("奇數索引總和：", odd_sum)
````

## 5. 函式實作：簡易計算機

````python
def calculator(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "除數不能是 0"
        return a / b
    else:
        return "不支援的運算符號"


print(calculator(2, 5, "/"))
````

## 6. 函式實作：溫度轉換

轉換公式：

- 開爾文溫度：`K = °C + 273.15`
- 華氏溫度：`°F = °C × 9 / 5 + 32`

````python
def temperature_converter(celsius):
    kelvin = celsius + 273.15
    fahrenheit = celsius * 9 / 5 + 32
    return kelvin, fahrenheit


kelvin, fahrenheit = temperature_converter(100)
print("開爾文：", kelvin)
print("華氏：", fahrenheit)
````

## 7. 函式實作：兩數的絕對值差

````python
def absolute_difference(a, b):
    difference = a - b

    if difference >= 0:
        return difference
    else:
        return -difference


print(absolute_difference(1, 21))
````

也可以使用 Python 內建的 `abs()` 簡化函式：

```python
def absolute_difference(a, b):
    return abs(a - b)
```

## 8. 函式實作：平均值與標準差

母體標準差的計算步驟：

1. 計算平均值。
2. 計算每個數字與平均值的差。
3. 將差平方後加總。
4. 除以資料數量。
5. 對結果開根號。

````python
def list_statistics(values):
    average = sum(values) / len(values)

    squared_difference_sum = 0
    for value in values:
        squared_difference_sum += (value - average) ** 2

    variance = squared_difference_sum / len(values)
    standard_deviation = variance ** 0.5

    return average, standard_deviation
````

````python
average, standard_deviation = list_statistics([1, 2, 3, 4, 5])
print("平均值：", average)
print("標準差：", standard_deviation)
````

## 9. 移除相鄰的重複元素

### 方法一：比較目前元素與前一個元素

````python
def compress_adjacent_by_index(values):
    result = [values[0]]

    for i in range(1, len(values)):
        if values[i] != values[i - 1]:
            result.append(values[i])

    return result
````

### 方法二：比較目前元素與結果串列的最後一個元素

````python
def compress_adjacent(values):
    result = []

    for value in values:
        if len(result) == 0 or value != result[-1]:
            result.append(value)

    return result
````

````python
numbers = [1, 1, 1, 3, 3, 2, 2, 5]

print(compress_adjacent_by_index(numbers))
print(compress_adjacent(numbers))
````

## 10. 綜合實作：猜數字遊戲

````python
answer = 37
count = 0

while True:
    guess = int(input("請輸入猜測的數字："))
    count += 1

    if guess < answer:
        print("too low")
    elif guess > answer:
        print("too high")
    else:
        print("correct")
        break

print("猜測次數：", count)
````

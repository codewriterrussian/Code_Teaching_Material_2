## 1. `print()` 與 `sep`

### 學習目標
理解 `print()` 可以一次輸出多個資料，而 `sep` 可以控制資料中間的分隔符號。

````python
print(1, 2, 3, 4, 5, sep=',')
print('台北', '台中', '高雄', sep=' -> ')
````

### 生活化例子：捷運路線
`sep=' -> '` 可以用來表示「路線」或「順序」。

````python
print('台北車站', '忠孝復興', '市政府', sep=' -> ')
````

### 小練習
請輸出三個地區的溫度，例如：`台北 22`、`台中 23`、`高雄 24`，並用逗號或箭頭分隔。

````python
print('台北 22', '台中 23', '高雄 24', sep=' | ')
print('2026', '5', '15', sep='/')
````

## 2. `input()`：讓使用者輸入資料

### 學習目標
理解 `input()` 可以讓程式從使用者取得資料。

````python
input('請輸入你的名字：')
````

````python
print(input('請輸入你的名字：'))
````

### 把輸入存成變數
如果之後還要使用輸入的資料，就要把它存進變數。

````python
name = input('請輸入你的名字：')
print('早安', name)
````

### 重點整理
- `input()` 會等待使用者輸入資料。
- `input()` 得到的資料預設都是 **字串 `str`**。
- 只寫 `input()` 會讀資料，但不方便後面繼續使用。
- 用變數保存後，後面可以重複使用。

## 3. `.split()`：把一行輸入切成多份

### 學習目標
理解一行輸入可能包含多個資料，可以用 `.split()` 拆開。

### 例子 1：用點號 `.` 拆生日
輸入範例：`1990.12.31`

````python
year, month, day = input('請輸入你的生日（格式：年.月.日）：').split('.')
print(year)
print(month)
print(day)
````

### 例子 2：用斜線 `/` 拆三個溫度
輸入範例：`25/30/40`

````python
temp1, temp2, temp3 = input('請輸入三個數字（格式：數字1/數字2/數字3）：').split('/')
print(temp1)
print(temp2)
print(temp3)
````

### 例子 3：用空白拆資料
`.split()` 不放東西時，預設用空白切割。

輸入範例：`Tom 13 Taipei`

````python
name, age, city = input('請輸入名字 年齡 城市：').split()
print(name)
print(age)
print(city)
````

### 容易出錯的地方
左邊變數數量要和右邊切出來的資料數量一致。

例如：`name, age, city = input().split()` 需要剛好輸入三個資料。

````python
data = input('請輸入一些資料，中間用空白隔開：').split()
print(data)
print(type(data))
````

## 4. 變數 Variable

### 學習目標
理解變數是用來儲存資料的名稱，也理解變數可以被新的資料覆蓋。

````python
name = '小明'
print(name)
````

````python
name = '小華'
print(name)
````

### 生活化例子：抽屜
變數像一個抽屜，抽屜裡一次主要放目前最新的東西。

````python
drawer = '珍奶'
print(drawer)

drawer = '紅茶'
print(drawer)
````

### 重點整理
- `name = '小明'`：把 `'小明'` 存進 `name`。
- 後面如果寫 `name = '小華'`，新的資料會覆蓋舊資料。
- 程式最後看到的是最新的值。

## 5. 變數與計算

### 學習目標
理解變數可以參與計算，也可以把計算結果存回原本的變數。

````python
x = 10
print(x + 10)
print(x)
````

上面的 `x + 10` 只是拿 `x` 去計算，**不會自動改變 `x`**。

````python
x = x + 10
print(x)
````

### 生活化例子：儲值卡加值
舊資料 → 計算 → 覆蓋成新資料。

````python
money = 10
money = money + 10
print(money)
````

## 6. 資料型態 Type

### 學習目標
認識常見資料型態：整數、浮點數、字串、布林值、串列。

| 型態 | Python 名稱 | 例子 | 說明 |
|---|---|---|---|
| 整數 | `int` | `10` | 沒有小數點的數字 |
| 浮點數 | `float` | `10.0` | 有小數點的數字 |
| 字串 | `str` | `'10'` | 文字資料 |
| 布林值 | `bool` | `True`, `False` | 真或假 |
| 串列 | `list` | `[1, 2, 3]` | 可以放多個資料的容器 |

````python
print(type(10))
print(type(10.0))
print(type('10'))
print(type(True))
print(type([10]))
````

### 修正提醒
Python 的布林值要寫成 `True` / `False`，第一個字母要大寫。

錯誤寫法：`x = true`  
正確寫法：`x = True`

````python
x = True
print(x)
print(type(x))
````

## 7. 型態轉換

### 學習目標
理解 `input()` 得到的資料通常是字串，如果要計算，常常需要轉成數字。

````python
temp = '28'
print(type(temp))
print(2 * float(temp))
print(2 * int(temp))
print(type(2 * float(temp)))
print(type(2 * int(temp)))
````

### 生活化例子：飲料總價
飲料一杯 60 元，使用者輸入買幾杯。

````python
price = 60
count = input('請輸入杯數：')
total = price * int(count)
print(total)
````

````python
price = 45
count = '3'
total = price * int(count)
print(total)
````

## 8. 布林值 Boolean

### 學習目標
理解布林值只有兩種：`True` 和 `False`。

````python
selection_bool = True
print(selection_bool)

selection_2_bool = False
print(selection_2_bool)
````

### 數字轉布林值
- 非零數字通常是 `True`
- `0` 是 `False`

````python
selection = 0.01
print(bool(selection))

selection_2 = 0
print(bool(selection_2))
````

### 字串轉布林值
- 空字串 `''` 是 `False`
- 非空字串，例如 `'0.01'`、`'False'`，都是 `True`

````python
selection_str = '0.01'
print(bool(selection_str))

selection_2_str = ''
print(bool(selection_2_str))

print(bool('False'))
````
